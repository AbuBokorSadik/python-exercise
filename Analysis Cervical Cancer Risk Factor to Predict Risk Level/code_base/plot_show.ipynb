```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/risk_factors_cervical_cancer.csv')
```

```python
df_heder=list(df)
for i in range(len(df_heder)-4):
    get_data=df[df_heder[i]].tolist()
    get_data_uni = list(df[df_heder[i]].unique())
    get_data_uni_count = {i:get_data.count(i) for i in get_data}
    fig = plt.figure()
    fig.set_size_inches(8, 5)
    ax = fig.add_subplot(111)
    for x in range(len(get_data_uni)):
        ax.bar(get_data_uni[x], get_data_uni_count[get_data_uni[x]])
    plt.title(df_heder[i])
    plt.xlabel('Unique Value of '+df_heder[i])
    plt.ylabel(df_heder[i])
    plt.savefig('../output/plot/pre/'+str(i))
    plt.show()
    plt.close()
```
