```python
import csv
import pandas as pd
```

```python
df = pd.read_csv('../output/risk_factors_cervical_cancer(formated).csv')
# df.head()
```

```python
# new target
ls=list(df['Hinselmann']+df['Schiller']+df['Citology']+df['Biopsy'])
print(ls)
```

```python
# make csv with new target value
df_heder=list(df)
df_heder=df_heder[0:32]
df_heder.append('target')

temp_dict=dict.fromkeys(df_heder)
for i in range(len(df_heder)-1):
    temp_dict[df_heder[i]]=list(df[df_heder[i]])
    
temp_dict['target']=ls
```

```python
temp_df=pd.DataFrame.from_dict(data=temp_dict, orient='index').T
temp_df.to_csv('../output/risk_factors_cervical_cancer(training).csv',index=False)
# temp_df.head()
```
