```python
import csv
import pandas as pd
import numpy as np
from scipy import stats
from statistics import mode
```

```python
df = pd.read_csv('../input/risk_factors_cervical_cancer.csv')
```

```python
df.head()
```

```python
# messing value replase
df_heder=list(df)
get_data=[]
replace_data=[]
temp_dict=dict.fromkeys(df_heder)
    
for i in range(len(df_heder)):
    counter=0
    get_data=df[df_heder[i]].tolist()
    temp_data=list.copy(get_data)
    replace_data=list.copy(get_data)
    if type(temp_data[0])==str:
        #string(integer) value in dataset
        if i!=5 and i!=6 and i!=8 and i!=10:
            temp_data = list(filter(lambda x: x!= '?', temp_data))
            temp_data_mode=mode(temp_data)
            for j in range(len(get_data)):
                if get_data[j]=='?':
                   replace_data[j]=str(temp_data_mode)                        
        #string(float) value in dataset column(5,6,8,10)
        else:
            temp_data = list(filter(lambda x: x!= '?', temp_data))
            temp_data = [float(i) for i in temp_data]
            temp_data_mean=np.mean(temp_data)
            for j in range(len(get_data)):
                if get_data[j]=='?':
                       replace_data[j]=str(temp_data_mean)

    temp_dict[df_heder[i]]=replace_data
```

```python
temp_df=pd.DataFrame.from_dict(data=temp_dict, orient='index').T
temp_df.head()
temp_df.to_csv('../output/risk_factors_cervical_cancer(formated).csv',index=False)
```
